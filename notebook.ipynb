## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana persebaran 10 kota dengan pelanggan terbanyak ?

```python
city_counts = customers_df['customer_city'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette="coolwarm")
plt.title('10 Kota dengan Jumlah Pelanggan Terbanyak', fontsize=16)
plt.xlabel('Kota', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2: Bagaimana distribusi besaran pelanggan bedasarkan wilayah ?


```python
state_counts = customers_df['customer_state'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette="magma")
plt.title('Distribusi Pelanggan Berdasarkan Wilayah', fontsize=16)
plt.xlabel('Wilayah (customer_state)', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python

product_category_counts = product_df['product_category_name'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=product_category_counts.index, y=product_category_counts.values)
plt.title('Distribusi Produk per Kategori')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Produk')
plt.xticks(rotation=45)
plt.show()
```
